### 1) Logistic Regression

#### 1.1 Equivalence of ERM and probabilistic approaches

\begin{array}{l}w_{\text {ERM }}=\operatorname{argmin}_{w} \frac{1}{n} \sum_{i=1}^{n} l_{\text{logistic}}(y, w) \\ w_{E R M}=\operatorname{argmin}_{w} \sum_{i=1}^{n} \log \left(1+\exp \left(-y^{(i)} w^{T} x^{(i)}\right)\right) ...... (1)\end{array}

$$p(y=1 | x ; w)=\frac{1}{1+\exp \left(-x^{\top} w\right)}$$

\begin{aligned} p(y=-1 | x ; w) &=1-p(y=1 | x ; w) \\ &=\frac{\exp \left(-x^{\top} w\right)}{1+\exp \left(-x^{\top} w\right)} \\ &=\frac{1}{1+\operatorname{exp}\left(x^{\top} w\right)}\end{aligned}

$$\Rightarrow p\left(y^{(i)} | x^{(i)} ; w\right)=\frac{1}{1+\exp \left(-y^{(i)} w^{\top} x^{(i)}\right)}$$ 

\begin{aligned} w_{\text {NLL }} &=\arg \min _{w} \sum_{i=1}^{n}-\log p\left(y^{(i)} | x^{(i)}, w\right) \\ &=\arg \min _{w} \sum_{i=1}^{n}-\log \frac{1}{1+\exp \left(-y^{(i)} w^{\top} x^{(i)}\right)} \\ &=\arg \min _{w} \sum_{i=1}^{n} \log \left(1+\exp \left(-y^{(i)} w^{\top} x^{(i)}\right)\right) ...... (2)\end{aligned}

$$(1), (2) \Rightarrow w_{\text{ERM}} = w_{\text{NLL}}$$

#### 1.2 - 1 Linearly Separable Data

\begin{aligned} \log \frac{p(y=1 | x^{(i)})}{p(y=-1 | x^{(i)})} &=\log \frac{\left(\frac{1}{1+\exp \left(-w^{T} x^{(i)}\right)}\right)}{\left(\frac{\exp \left(-w^{\top} x^{(i)}\right)}{1+\exp \left(-w^{\top} x^{(i)}\right)}\right)} \\ &=\log \exp \left(w^{\top} x^{(i)}\right) \\ &=w^{\top} x^{(i)} \end{aligned}

At the separation boundary, 
\begin{aligned} & p(y=1 | x)=p(y=-1 | x) \\ \Rightarrow \quad & \log \frac{p(y=1 | x)}{p(y=-1 | x)} \\ \Rightarrow \quad & w^{\top} x=0 \end{aligned}

This is linear equation in w hence the boundary is linear.

#### 1.2 - 2 Linearly Separable Data

All points are correctly classified 
\begin{array}{l}\Rightarrow y^{(i)}=\operatorname{sign}\left(w^{\top} x^{(i)}\right) \\ \Rightarrow y^{(i)} w^{\top} x^{(i)}=\left|w^{\top} x^{(i)}\right|\end{array}
for all i <br>

Let $c_1, c_2 \in \mathbb{R}$ such that $c_2 \gt c_1$ <br>

\begin{aligned} N L L\left(c_{1} w\right) &=\sum_{i=1}^{n} \log \left(1+\operatorname{exp}\left(-y^{(i)}\left(c_{1} w^{T}\right) x^{(i)}\right)\right) \\ &=\sum_{i=1}^{n} \log \left(1+\exp \left(-c_{1} \cdot\left(y^{(i)} w^{T} x^{(i)}\right)\right)\right) \\ &=\sum_{i=1}^{n} \log \left(1+\exp \left(-c_{1}\left|w^{\top} x^{(i)}\right|\right)\right)\end{aligned}
\begin{array}{l}\operatorname{NLL}\left(c_{2} w\right)=\sum_{i=1}^{n} \log \left(1+\operatorname{exp}\left(-c_{2}\left|w^{\top} x^{(i)}\right|\right)\right)\end{array}

\begin{aligned} c_{2}>c_{1} & \Rightarrow \quad-c_{2}<-c_{1} \\ & \Rightarrow \quad-c_{2}\left|w^{\top} x^{(i)}\right|<-c_{1}\left|w^{\top} x^{(i)}\right| \\ & \Rightarrow \quad \exp \left(-c_{2}\left|w^{\top} x^{(i)}\right|\right)<\exp \left(-c_{1}\left|w^{\top} x^{(i)}\right|\right) \\ & \Rightarrow \log \left(1+\exp \left(-c_{2}\left|w^{\top} x^{(i)}\right|\right)\right)<\log \left(1+\exp \left(-c_{1}\left|w^{\top} x^{(i)}\right|\right)\right) \\ & \Rightarrow \sum_{i=1}^{n} \log \left(1+\exp \left(-c_{2}\left|w^{\top} x^{(i)}\right|\right)\right)<\sum_{i=1}^{n} \log \left(1+\exp \left(-c_{1}\left|w^{\top} x^{(i)}\right|\right)\right) \\ & \Rightarrow \quad \operatorname{NLL}\left(c_{2} w\right) \quad<\quad \operatorname{NLL}\left(c_{1} w\right) \end{aligned}

\begin{array}{l}\Rightarrow \forall c_{2}>c_{1}, \quad \text { NLL }\left(c_{2} w\right)<N L L(c_1 w) \\ \Rightarrow \text { NLL can be made arbitrarily small by increasing } c \\ \Rightarrow \text { likelihood can be made arbitrarily large by increasing } c\end{array}

#### 1.3 - 1 Regularized Logistic Regression

$$J(w)=\frac{1}{n} \sum_{k=1}^{n} \log \left(1+\exp \left(-y^{(k)} w^{\top} x^{(k)}\right)\right)+\lambda\|w\|^{2}$$

$$\frac{\partial J}{\partial w_{i}}=\frac{1}{n} \sum_{k=1}^{m} \frac{1}{1+\exp \left(-y^{(k)} \omega^{\top} x^{(k)}\right)} \cdot \exp \left(-y^{(k)} w^{\top} x^{(k)}\right) \cdot\left(-y^{(k)} x_{i}^{(k)}\right)+2 \lambda\left(w_{1}+\cdots+w_{d}\right)$$

$$\frac{\partial J}{\partial w_{i}}=\frac{1}{n} \sum_{k=1}^{n} \frac{-y^{(k)} x_{i}^{(k)}}{1+\exp \left(y^{(k)} w^{\top} x^{(k)}\right)}+2 \lambda\left(w_{1}+\ldots+w_{d}\right)$$

$$\frac{\partial J}{\partial w_{i} \partial \omega_{j}}=\frac{1}{n} \sum_{k=1}^{n}\left(-y^{(k)} x_{i}^{(k)}\right) \frac{-1}{\left(1+\exp \left(y^{(k)} w^{\top} x^{(k)}\right)\right)^{2}} \cdot \exp \left(y^{(k)} w^{\top} x^{(k)}\right) \cdot\left(y^{(k)} x_{j}^{(k)}\right)+2 \lambda$$

The i,j entry of the Hessian of J is given by:

$$H_{i j}=\frac{1}{n} \sum_{k=1}^{n} \frac{\left(y^{(k)}\right)^{2} \exp \left(y^{(k)} w^{\top} x^{(k)}\right)}{\left(1+\exp \left(y^{(k)} w^{\top} x^{(k)}\right)\right)^{2}} \cdot x_{i}^{(k)} x_{j}^{(k)}+2 \lambda$$

$$\text { Let } f(k)=\frac{\left(y^{(k)}\right)^2 \exp \left(y^{(k)} w^{\top} x^{(k)}\right)}{\left(1+\exp \left(y^{(k)} w^{\top} x^{(k)}\right)\right)^{2}} \cdot \frac{1}{n}$$
$$\text { Note that } f(k) \geq 0$$

$$H=\left(\begin{array}{ccc}\sum_{k=1}^{n} f(k) \cdot x_{1}^{(k)} \cdot x_{1}^{(k)}+2 \lambda & \ldots & \sum_{k=1}^{n} f(k) \cdot x_{1}^{(k)} \cdot x_{d}^{(k)}+2 \lambda \\ \vdots & \\ \sum_{k=1}^{n} f(k) \cdot x_{d}^{(n)} \cdot x_{1}^{(k)}+2 \lambda & \ldots & \sum_{k=1}^{n} f(k) \cdot x_{d}^{(k)} \cdot x_{d}^{(k)}+2 \lambda\end{array}\right)$$

$$H=\sum_{k=1}^{n} f(k)\left(\begin{array}{cc}x_{1}^{(k)} \cdot x_{1}^{(k)} & x_{1}^{(k)} \cdot x_{d}^{(k)} \\ \vdots & \\ x_{d}^{(k)} \cdot x_{1}^{(k)} & x_{d}^{(k)} \cdot x_{d}^{(k)}\end{array}\right) + 2 \lambda I_{d}$$

$$H =\sum_{k=1}^{n} f(k) x^{(k)}\left(x^{(k)}\right)^{\top}+2 \lambda I_{d}$$

$$\text { Let } z \in \mathbb{R}^{d}$$

\begin{aligned} z^{\top} H z &=z^{\top}\left(\sum_{k=1}^{n} f(k) x^{(k)}\left(x^{(k)}\right)^{\top}\right) z+2 \lambda z^{\top} z \\ &=\sum_{k=1}^{n} f(k) \cdot z^{\top} x^{(k)}\left(x^{(k)}\right)^{\top} z+2 \lambda z^{\top} z \\ &=\sum_{k=1}^{n} f(k) \cdot\left(\left(x^{(k)}\right)^{\top} z\right)^{\top}\left(\left(x^{(k)}\right)^{\top} z\right)+2 \lambda z^{\top} z \\ &=\sum_{k=1}^{n} f(k)\left\|\left(x^{(k)}\right)^{\top} z\right\|^{2}+2 \lambda\left\|_{z}\right\|^{2} \\ & \geq 0 \end{aligned}

\begin{array}{l}\Rightarrow H \text { is positive semidefinite } \forall w \in \mathbb{R}^{d} \\ \Rightarrow J \text { is convex }\end{array}


#### 1.3 - 2 Regularized Logistic Regression

```python
import numpy as np
import pandas as pd
from scipy import optimize as optimize

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X_train = pd.read_csv('../data/logistic_regression/X_train.txt', header = None).to_numpy()
y_train = pd.read_csv('../data/logistic_regression/y_train.txt', header = None).to_numpy()
X_val = pd.read_csv('../data/logistic_regression/X_val.txt', header = None).to_numpy()
y_val = pd.read_csv('../data/logistic_regression/y_val.txt', header = None).to_numpy()
```

```python
def preprocess(X):
    rows = X.shape[0]
    cols = X.shape[1]

    # normalize
    for j in range(cols):
        col = X[:, j]
        X[:, j] = (col - np.mean(col)) / np.std(col)

    # add bias column
    X = np.append(X, np.ones(rows).reshape(rows, 1), 1)

    return X
```

```python
X_train = preprocess(X_train)
X_val = preprocess(X_val)
```

```python
def log_likelihood(theta, X, y, l2_param = 1):
    n = X.shape[0]
    loss = 0

    for i in range(n):
        loss += np.logaddexp(0, -y[i] * np.dot(X[i], theta))
    
    loss /= n

    return loss


def f_objective(theta, X, y, l2_param=1):
    '''
    Args:
        theta: 1D numpy array of size num_features
        X: 2D numpy array of size (num_instances, num_features)
        y: 1D numpy array of size num_instances
        l2_param: regularization parameter

    Returns:
        objective: scalar value of objective function
    '''

    loss = log_likelihood(theta, X, y, l2_param) + l2_param * np.linalg.norm(theta)

    return loss
```

#### 1.3 - 3 Regularized Logistic Regression

```python
def fit_logistic_reg(X, y, objective_function, l2_param=1):
    '''
    Args:
        X: 2D numpy array of size (num_instances, num_features)
        y: 1D numpy array of size num_instances
        objective_function: function returning the value of the objective
        l2_param: regularization parameter

    Returns:
        optimal_theta: 1D numpy array of size num_features
    '''

    theta = np.ones(X.shape[1])
    opt = optimize.minimize(objective_function, x0 = theta, args = (X, y, l2_param))

    return opt.x
```

```python
theta = fit_logistic_reg(X_train, y_train, f_objective, 10**-2)
```

```python
error_map = {}
```

#### 1.3 - 4 Regularized Logistic Regression

```python
# LAMBDA_LIST = [10**-12, 10**-11, 10**-10.75, 10**-10.5, 10**-10.25, 10**-10, 10**-9.5, 10**-9]
# LAMBDA_LIST = [10**-16, 10**-15, 10**-14, 10**-13, 10**-12, 10**-11, 10**-10]
LAMBDA_LIST = [10**-12.5, 10**-12.4, 10**-12.3, 10**-12.2, 10**-12.1, 10**-11.9, 10**-11.8, 10**-11.7, 10**-11.6]
errors = []

for l2_param in LAMBDA_LIST:
    if l2_param not in error_map:
        theta_opt = fit_logistic_reg(X_train, y_train, f_objective, l2_param)
        val_error = log_likelihood(theta_opt, X_val, y_val, l2_param)
        errors.append(val_error)
        error_map[l2_param] = val_error
    else:
        val_error = error_map[l2_param]
        errors.append(val_error)
```

```python
lambda_all = sorted(error_map.keys())
errors_all = []

for lambda_l in lambda_all:
    errors_all.append(error_map[lambda_l])
```

```python
fig, ax = plt.subplots(figsize = (8,6))
ax.grid()
ax.set_title("Validation Performance vs L2 Regularization")
ax.set_xlabel("L2-Penalty Regularization Parameter")
ax.set_ylabel("Mean Squared Error")
ax.semilogx(lambda_all, errors_all)
```

The best validation loss is observed for $\lambda = 10^{-12}$

### 3) Coin Flipping with Partial Observability

### Question 1

\begin{aligned} p\left(x=H | \theta_{1}, \theta_{2}\right) &=p\left(x=H | z=H, \theta_{1}, \theta_{2}\right) p\left(z=H | \theta_{1}, \theta_{2}\right)+p\left(x=H | z=1, \theta_{1}, \theta_{2}\right) p\left(z=T | \theta_{1}, \theta_{2}\right) \\ &=\theta_{2} \theta_{1}+0 \\ &=\theta_{1} \theta_{2} \end{aligned}

### Question 2

Let #{A} denote the number of times an event A has occurred. <br>
Let 
\begin{array}{l}\#\{z=H, x=H\}=a \\ \#\{z=H, x=T\}=b \\ \#\{z=T, x=T\}=c \\ \#\{z=T, x=H\}=d\end{array}
To use MLE, we need to know the above four quantities. <br>

\begin{aligned} & p(x=T | z=T)=1 \\ & \Rightarrow  \#\{z=T, x=H\}=0 \\ & \Rightarrow d=0 \end{aligned}
Also,
$$\#\{z=H, x=H\}+\#\{z=T, x=H\}=n_{h}$$
\begin{array}{l}\Rightarrow \quad a+d=n_{h} \\ \Rightarrow \quad a=n_{h}\end{array}
\begin{aligned} & \#\{z=H, x=T\}+\#\{z=T, x=T\}=n_{t} &\end{aligned}
$$\Rightarrow  b+c =n_{t} $$

With this information, we can only compute a, d but not b and c. <br>
All we can say about b,c is: $b + c = n_t$ <br>
Therefore, we cannot use MLE to estimate the parameters.

### Question 3

Continuing the notation from Question 2,
\begin{array}{l}a+b=c_{h} \\ c+d=c_{t} \\ a+d=n_{h} \\ b+c=n_{t}\end{array}
Solving these equations, we get, 
\begin{array}{l}a=n_{n} \\ c=c_{t} \\ b=c_{h}-n_{h}=n_{t}-c_{t} \\ d=0\end{array}
